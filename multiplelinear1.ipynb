```python
# checking for outliers in numerical columns
plt.figure(figsize = (15,25))   #figsize -- allows us to specify the width and height of a figure in unit inches
count = 1
for col in df.columns:
    if(df[col].dtype != 'O'):
        plt.subplot(7,2,count) #subplot(nrows,ncolumns)
        plt.boxplot(df[col])   #gives us information about min value
        plt.title(col)
        count +=1
              
plt.show()
```

```python
#removing outliers
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
for col in df.columns:
    if(col!='CHAS'):
        df = remove_outlier(df,col)
```

```python
from sklearn.model_selection import train_test_split#splitting data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2 , random_state = 0)
```

```python
#model selection
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#fitting data 
model.fit(X_train,y_train)
```

```python
#Predictions
y_pred= model.predict(X_test)
```

```python
#checking the r2score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
